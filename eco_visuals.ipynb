### Graphing Ecosystem Data

```python
import ecosystem
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

Running the ecosystem

```python
ecosystem.reset_pops()

ecosystem.run(invasive=False)
eco = ecosystem.eco_data

ecosystem.run(invasive=True)
eco_inv = ecosystem.eco_data
```

Getting needed data

```python
cod_pop = np.array(eco['cod_pop'])
cod_pop_inv = np.array(eco_inv['cod_pop'])
cod_pop_inv[0] = 10000 # Adjustment
x = eco['step']
```

Plotting the data

```python
plt.plot(x, cod_pop, color='blue')
plt.plot(x, cod_pop_inv, color='red')
plt.title("Cod Population Over 5 Years")
plt.xlabel("Day")
plt.ylabel("Cod Population")
plt.legend(["No Invasive", "Invasive"])
plt.grid(alpha=.25)
```

Plotting Growth and Death data

```python
# Growth
cod_g = np.array(eco['cod_growth'])
cod_g_inv = np.array(eco_inv['cod_growth'])

# Death
cod_d = np.array(eco['cod_death'])
cod_d_inv = np.array(eco_inv['cod_death'])

x = eco['step']

# Plot
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(x, cod_g, color='blue')
ax.plot(x, cod_g_inv, color='red')
ax.plot(x, cod_d, color='green')
ax.plot(x, cod_d_inv, color='orange')
ax.set_title("Cod Growth & Death Metrics Over 5 Years")
ax.set_xlabel("Day")
ax.set_ylabel("Number of Cod")
ax.legend(["Growth (No Invasive)", "Growth (Invasive)", "Death (No Invasive)", "Death (Invasive)"], framealpha=1, loc="best")
ax.grid(alpha=.25)
```
