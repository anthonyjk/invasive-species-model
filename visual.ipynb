```python
import ecosystem
import matplotlib.pyplot as plt
```

```python
ecosystem.run()
```

```python
plt.figure(figsize=(12, 6))

fig, ax = plt.subplots()
xa = ax.twinx()
#xa.plot(ecosystem.x, ecosystem.k, label="Krill")
ax.plot(ecosystem.x, ecosystem.p, label="Penguins", color='black')
xa.plot(ecosystem.x, ecosystem.c, label="Cod", color='orange')
plt.xlabel("Day")
plt.ylabel("Population")
ax.legend(['penguin'], loc='upper right')
xa.legend(['cod'])
plt.show()
```
