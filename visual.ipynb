```python
import ecosystem
import matplotlib.pyplot as plt
```

```python
ecosystem.run()
no_inv = ecosystem.eco_data

ecosystem.reset_pops()

ecosystem.run(invasive=True)
inv = ecosystem.eco_data
```

```python
plt.plot(no_inv['step'], no_inv['cod_pop'], color='red')
plt.plot(inv['step'], inv['cod_pop'], color='blue')
```

```python
import random
import numpy as np
cod_itr = []
for i in range(50):
    ecosystem.reset_pops()
    cbr = random.uniform(.75, 2.25)
    #cdr = random.uniform(0.01, 0.3)
    cdr=0.1
    ecosystem.run(cod_birth_rate=cbr, cod_death_rate=cdr)
    cod_itr.append([cbr, cdr, sum(ecosystem.eco_data['cod_pop']), 0])

    ecosystem.reset_pops()
    ecosystem.run(invasive=True,cod_birth_rate=cbr, cod_death_rate=cdr)
    cod_itr.append([cbr, cdr, sum(ecosystem.eco_data['cod_pop']), 1])
```

```python
cod_itr = np.array(cod_itr)

cmap = np.where(cod_itr[:,3] == 1, 'red', 'blue')

plt.scatter(cod_itr[:,0], cod_itr[:,2], color=cmap)
```

```python
cod_itr = np.array(cod_itr)

cmap = np.where(cod_itr[:,3] == 1, 'red', 'blue')

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(cod_itr[:,0], cod_itr[:,2], np.sum(cod_itr[:,(0,2)] ** 2, axis=1), color = cmap, s = 50)
ax.view_init(azim=20, elev=25)
plt.title("simple 3D scatter plot")
```
